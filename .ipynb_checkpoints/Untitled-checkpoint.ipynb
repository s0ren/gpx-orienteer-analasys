````python
import gpxpy
import gpxpy.gpx
````

````python
gpx_file = open('data/activity_4931694757.gpx', 'r')
gpx = gpxpy.parse(gpx_file)
````

````python
for track in gpx.tracks:
    for segment in track.segments:
        for point in segment.points:
            print('Point at ({0},{1}) -> {2}'.format(point.latitude, point.longitude, point.elevation))
````

````python
import xmltodict
````

````python
tcx_file = open('data/activity_4931694757.tcx', 'r')
tcx = xmltodict.parse(tcx_file.read())
#tcx
````

````python
laps = tcx['TrainingCenterDatabase']['Activities']['Activity']['Lap']

count = 0

for lap in laps:
    print("Ben {0}\t time: {1}\t distance: {2}\t avgTempo: {3} (sec/m) ".format(count, 
                        lap['TotalTimeSeconds'], lap['DistanceMeters'],
                       float(lap['TotalTimeSeconds']) / float(lap['DistanceMeters'])))                               
    count += 1
    
````

````python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
mu = 0
std = 1
x = np.linspace(start=-4, stop=4, num=100)
y = stats.norm.pdf(x, mu, std) 
plt.plot(x, y)
plt.show()
````
